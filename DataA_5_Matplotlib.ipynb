#Importing Modules

```
import matplotlib.pyplot as plt
import numpy as np
```

#Plotting an array

```
arr = np.random.randn(10)
arr
```

```
plt.plot(arr)
plt.show()
```

#Plotting with Labels

```
plt.plot(arr)

plt.title('First plot')
plt.xlabel('Indices')
plt.ylabel('Values')
plt.show()
```

#Playing around with the plottings

```
plt.plot(arr, '--k')  #'go'
plt.title('First plot')
plt.xlabel('Indices')
plt.ylabel('Values')
plt.show()
```

#Bar Graphs

```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
plt.bar(divisions, avg_marks, color='#00ff00')
plt.title('Bar Graph')
plt.xlabel('Division')
plt.ylabel('Marks')
plt.show()
```

```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
plt.barh(divisions, avg_marks, color='#00ff00')
plt.title('Bar Graph')
plt.xlabel('Division')
plt.ylabel('Marks')
plt.show()
```

#Multi-bar Graphs

---



```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
boys_avg_marks = [78,96,53,71,82]
girls_avg_marks = [88,76,99,71,62]
index = np.arange(5)
width = .3
plt.bar(index,avg_marks,width, color='blue', label = 'Division Marks')
plt.bar(index+width,boys_avg_marks,width, color='green', label = 'Boys Marks')
plt.bar(index+2*width,girls_avg_marks,width, color='red', label = 'Girls Marks')
plt.xlabel('Div')
plt.ylabel('Marks')
plt.xticks(index+width,divisions)
plt.legend()
plt.show()
```

#Stacked Graphs

```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
boys_avg_marks = [78,96,53,71,82]
girls_avg_marks = [88,76,99,71,62]
index = np.arange(5)
plt.figure(figsize= (7, 7))
width = .3
plt.bar(index,avg_marks,width, color='blue', label = 'Division Marks')
plt.bar(index,boys_avg_marks,width, color='green', label = 'Boys Marks', bottom = avg_marks)
plt.bar(index,girls_avg_marks,width, color='red', label = 'Girls Marks', bottom = boys_avg_marks)
plt.xlabel('Div')
plt.ylabel('Marks')
plt.xticks(index,divisions)
plt.legend(title = 'Avg',loc = 'upper right')
plt.show()
```

#Pie Charts

```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
plt.pie(avg_marks, labels= divisions )


plt.show()
```

```
divisions = ['A', 'B', 'C','D','E']
avg_marks = [40,70,87,50,60]
Explode = [0,0.2,0,0,.3]
plt.title('Class avg', loc = 'right')
plt.pie(avg_marks, labels= divisions,autopct='%1.1f%%',  explode = Explode, shadow= True)
plt.show()
```

#Histogram

```
x = np.random.randn(1000)
plt.title('Histogram')
plt.xlabel('Random Data')
plt.ylabel('Frequency')
plt.hist(x)
plt.show()
```

#Scatter Plots and 3-D plotting

```
height = np.random.randint(140,200, 10)
height.sort()
print(height)
weight = np.random.randint(60,100, 10)
weight.sort()
print(weight)
plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(height, weight)
plt.title('Scatter Plot')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

#3d View

```
from mpl_toolkits import mplot3d
```

```
ax = plt.axes(projection= '3d')
ax.scatter3D(height, weight)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```

```
ax = plt.axes(projection= '3d')
ax.plot3D(height, weight)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```
